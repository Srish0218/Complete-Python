```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0 , 10)
y = np.arange(11 , 21)
```

```python
#scatter plot

plt.scatter(x , y , c = 'cyan')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot')
plt.savefig('Scatter1.png')
```

```python
y = x * x
```

```python
plt.plot(x , y , 'ro--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot()')
```

```python
#subplots
plt.subplot(2 , 3 , 1)
plt.plot(x , y, 'r--')

plt.subplot(2 , 3 , 2)
plt.plot(x , y, 'c*--')

plt.subplot(2 , 3 , 4)
plt.plot(x , y, 'go')

plt.subplot(2 , 3 , 6)
plt.plot(x , y, 'bo')
```

```python
x= np.arange(0 , 4* np.pi , 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2 , 2 , 1)
plt.title("Sin wave")
plt.plot(x, y1)

plt.subplot(2 , 2 , 2)
plt.title("Cos wave")
plt.plot(x, y2)
```
